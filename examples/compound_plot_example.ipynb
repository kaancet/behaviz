```python
import numpy as np
import behaviz as bv
from behaviz.spec import FigureSpec, AxisSpec, PlotSpec, ScaleType
from behaviz.manipulations import VisualManipulator

from bokeh.plotting import show
from bokeh.io import output_notebook
```

### Set renderer
Keep in mind bokeh needs an additional show() call to display the figure

```python
bv.set_renderer("bokeh")
```

### Define the plotting function

```python
SINAPLOT_SPEC = PlotSpec(
    figure=FigureSpec(figsize=(6,6), dpi=300),
    x=AxisSpec(label="Contrast", unit="%", scale=ScaleType.LINEAR),
    y=AxisSpec(label="Hit rate", unit="%", scale=ScaleType.LINEAR),
    title="SIna plot"
)

def sinaplot(x,ys,bin_width, **kwargs):
    
    f, ax, vp = bv.plot_violin(x,ys,spec=SINAPLOT_SPEC, alpha=0.3, color="#4D4C4C", **kwargs)
    
    
    bvm = VisualManipulator()
    
    for xi,yi in zip(x,ys): 
        mnp = bvm.jitter(xi,yi,kind="beeswarm",axis="x",bin_width=bin_width, side="center", width=0.2)
        f,ax = bv.plot_scatter(xi+mnp.x,mnp.y,ax=ax,spec=SINAPLOT_SPEC,color="#000000",**kwargs)
    
    return f,ax
```

```python
x = np.array([1,2,3,4,5])
ys = [np.random.randn(60),
      np.random.randn(70),
      np.random.randn(100),
      np.random.randn(50),
      np.random.randn(80),]

_,ax = sinaplot(x,ys,0.2)

# if renderer is bokeh
if bv.get_renderer().name == "bokeh":
    output_notebook()
    show(ax)
```
