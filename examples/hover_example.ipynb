```python
%matplotlib widget
import numpy as np
import behaviz as bv

from bokeh.plotting import show
from bokeh.io import output_notebook
```

### Set renderer
Keep in mind bokeh needs an additional show() call to display the figure

```python
bv.set_renderer("bokeh")
```

```python
x = np.array([1,2,3,4,5])
y = np.array([10,15,35,60,88])/100

_,ax = bv.plot_line(x,y,label="data",color="#349888",hover_annotate=True,hover_labels=("x-value","y-value!"))

if bv.get_renderer().name == "bokeh":
    output_notebook()
    show(ax)
```
