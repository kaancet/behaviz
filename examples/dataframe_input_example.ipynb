```python
%matplotlib widget
import numpy as np
import polars as pl
import behaviz as bv
from behaviz.spec import FigureSpec, AxisSpec, PlotSpec, ScaleType

from bokeh.plotting import show
from bokeh.io import output_notebook
```

### Set renderer
Keep in mind bokeh needs an additional show() call to display the figure

```python
bv.set_renderer("seaborn")
```

```python
df = pl.DataFrame(data={"height":np.random.rand(50), "weight":np.random.rand(50)})
df.head()
```

```python
DEFAULT_SPEC = PlotSpec(
    figure=FigureSpec(figsize=(3.5, 3.5), dpi=300),
    x=AxisSpec(scale=ScaleType.LINEAR),
    y=AxisSpec(scale=ScaleType.LINEAR),
    show_legend=False,
)

_,ax = bv.plot_scatter(data=df,x="weight",y="height",label="data",color="#349888",hover_annotate=True, spec=DEFAULT_SPEC)

if bv.get_renderer().name == "bokeh":
    output_notebook()
    show(ax)
```
