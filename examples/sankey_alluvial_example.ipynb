# Sankey & Alluvial flow plots

Both take a **tidy flow table** — one row per flow, with `source`, `target`, `value` columns.

- **`plot_sankey`** — a single set of flows (source → target), two columns.
- **`plot_alluvial`** — a multi-stage sankey; add a `stage` column that orders the transitions.

Ribbons are coloured by their **source** node. Works the same on matplotlib, seaborn and bokeh.

````python
import pandas as pd
import behaviz as bv

bv.set_renderer("matplotlib")
````

## 1. Sankey — single-stage flows

A tidy table of `source, target, value`. Here: how visitors from each channel convert.

````python
flows = pd.DataFrame({
    "source": ["Ad", "Ad", "Search", "Search", "Direct", "Direct"],
    "target": ["Signup", "Bounce", "Signup", "Bounce", "Signup", "Bounce"],
    "value":  [120, 80, 200, 50, 90, 30],
})
flows
````

````python
fig, ax = bv.plot_sankey(flows)
````

### Custom colours

`palette=` accepts a `{node: colour}` dict (or a list). Nodes not named fall back to the default cycle.

````python
fig, ax = bv.plot_sankey(
    flows,
    palette={"Ad": "#e4572e", "Search": "#3366cc", "Direct": "#29a329"},
)
````

## 2. Alluvial — multi-stage flows

Same tidy table, plus a **`stage`** column naming each transition. Categories persist across
stages, so you see how membership moves over time. Here: a cohort across three weeks.

````python
alluvial = pd.DataFrame({
    "source": ["Active", "Active", "Idle",   "Idle",  "Churned",
               "Active", "Active", "Idle",   "Idle",  "Churned"],
    "target": ["Active", "Idle",   "Active", "Idle",  "Churned",
               "Active", "Idle",   "Active", "Churned", "Churned"],
    "value":  [60, 20, 15, 30, 40,
               55, 25, 20, 25, 55],
    "stage":  ["W1"] * 5 + ["W2"] * 5,
})
alluvial
````

````python
fig, ax = bv.plot_alluvial(alluvial, stage="stage")
````

**Stage ordering:** transitions are ordered by *first appearance* in the rows, not sorted.
If your `stage` values are numeric (`0, 1, 2`) or you want a specific order, sort the frame first:

```python
alluvial = alluvial.sort_values("stage")
```

The `stage` values only group and order the transitions — they aren't drawn (node labels come from source/target).

## 3. Same call, other backends

The identical call renders on seaborn (matplotlib under the hood) and bokeh.

````python
bv.set_renderer("seaborn")
fig, ax = bv.plot_sankey(flows)
````

````python
from bokeh.io import output_notebook, show

output_notebook()
bv.set_renderer("bokeh")
fig, ax = bv.plot_alluvial(alluvial, stage="stage")
show(fig)
````
