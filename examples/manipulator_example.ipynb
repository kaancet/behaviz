```python
import numpy as np
import behaviz as bv
from bokeh.plotting import show
from bokeh.io import output_notebook
```

### Set the renderer
Keep in mind bokeh needs an additional show() call to display the figure

```python
bv.set_renderer("seaborn")
```

### Visually manipulate the data

```python
x = np.arange(0, 5*np.pi, 0.1)
y = np.sin(x)

_,ax = bv.plot_line(x,y, label="data")

# add jitter to y

bvm = bv.manipulations.VisualManipulator(seed=42)
mnp = bvm.jitter(x,y,kind="uniform",axis="y",strength=0.5)

# plot jitter
_,ax = bv.plot_scatter(mnp.x,mnp.y,ax=ax,color="#D50202",label="jittered")

# smooth the jittered
mnp_smooth = bvm.smooth(mnp.x, mnp.y, kind="gaussian", window=5, axis="y")

# plot the jitter
_,ax = bv.plot_line(mnp_smooth.x,mnp_smooth.y, ax=ax,color="#01B348", label="smoothed jittered")

# if renderer is bokeh
if bv.get_renderer().name == "bokeh":
    output_notebook()
    show(ax)
```

### Normalise the data

```python
x = np.arange(0,100)
y = np.append(np.random.randn(len(x)//2), np.random.randn(len(x)//2)+10)

baseline_idx = np.arange(0,10)

_,ax = bv.plot_line(x,y, label="data")

bvm = bv.manipulations.VisualManipulator(seed=42)
mnp = bvm.normalise(x,y,kind="zscore",axis="y",baseline_idx=baseline_idx)

_,ax = bv.plot_line(mnp.x, mnp.y, ax=ax,color="#D50202", label=f"{mnp.metadata['kind']} normalised")

if bv.get_renderer().name == "bokeh":
    output_notebook()
    show(ax)
```

### Bin the data

```python
x = np.arange(0,100,5)
y = np.random.rand(len(x))
_,ax = bv.plot_line(x,y, label="data", marker="o", markersize=8)

bvm = bv.manipulations.VisualManipulator()
mnp = bvm.binning(x,y,kind="mean",axis="x")

print(mnp.x)
print(mnp.y)
_,ax = bv.plot_bar(mnp.x,mnp.y, width=0.9,ax=ax,color="#DA2424")

if bv.get_renderer().name == "bokeh":
    output_notebook()
    show(ax)
```
