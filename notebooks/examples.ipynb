```python
import numpy as np
from behaviz.psychometric import plot_psychometric
from behaviz.psychometric import PSYCHOMETRIC_SPEC

def sigmoid(x_arr, x0, L, k, p_false=0):
    """

    x0 = x value of the sigmoid midpoint
    L = max value of sigmoid
    k = logistic growth rate aka steepness
    b = y-offset(if needed)
    """
    return L / (1.0 + np.exp(-k * (x_arr - x0))) + p_false

x = np.array([1,2,3,4,5,6])
y = np.array([10,15,35,60,88,100])/100
err = np.array([5,5,5,10,20,5])/100

f = plot_psychometric(x,y,err,fit_func=sigmoid,spec=PSYCHOMETRIC_SPEC.with_size((3,2)).with_ylim(0,1.05), 
                      color="r",linewidth=0.0001, marker="*",markersize=10,linestyle="--",elinewidth=2,
                      fit_color="#889988",fit_linestyle=":",fit_linewidth=2)
```

### Combine plot types

```python
import numpy as np
from behaviz.core.core import plot_errorbar
from behaviz.impact import plot_impact
from behaviz.core.plot_setup import plot_function
from scipy import stats


def plot_impact_SEM(x,ys):
    
    x = np.asarray(x).ravel()
    ys = np.asarray(ys)
    
    f, ax = plot_impact(x,ys,jitter=0.02,dot_color="#E31616",color="r",dot_s=60)
    
    # Calculate the mean and standard error
    mean = np.mean(ys,axis=0)
    se = stats.sem(ys,axis=0)

    print(mean)
    print(se)    
    f, ax = plot_errorbar(x,mean,se,ax=ax, marker="o", markersize=20, elinewidth=3) 
    
    return f,ax


x = [1,2,3,4,]
ys = [[1,3,5,8],[2,6,3,8],[4,5,6,8],[5,7,2,8]]

f = plot_impact_SEM(x,ys)
```
